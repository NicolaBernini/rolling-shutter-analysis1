
# Rolling Shutter 

## Overview 

CMOS Sensor have different advantages like 

  * low cost 
  * good performance 
  
but also a major disadvantage 

  * rolling shutter 

<br/>

Rolling Shutter consists of *reading the sensor line by line*  

**Note** that while reading the image can change 

* the last line could not be anymore consistent with the first line 
* it could lead to **non consistent images** (absurd ones)

This phenomenon does not happen with CCD and Mechanical Shutters providing **Global Shutter**

* With G.S. the sensor is exposed and then "isolated" (can not change anymore because of scene changes) for a very small time 

<br/>

With Shutter the problem is **time** 

* Global Shutters expose the sensor for a time interval very small 
    * then it is possible to assume the scene as static (for the vast majority of the phenomena of interest) 
* Rolling Shutter could be too slow wrt some phenomena happening in the scene 
    * theh it is not possible to assume the scene as static anymore 

<br/> 

Hence it is possible to think at Rolling Shutter Camera as a device producing an image where each pixel should carry the extra information (beside color of course) of **timestamp** 


## Timestamping Pixels 

Let's consider the following Parameters for a Rolling Shutter Camera 

* $ t_{\tau} $ (constant) Time for a complete scan of the sensor 
  * It should be provided on the Sensor Datasheet 
* $ t_{id} $ (constant) Idle Time (when switching from one frame to another)
  * It should be provided on the Sensor Datasheet 
* $ u_{i} = \{ (r,c) \}_{i} (Bidimensional Array) $ the $ i $ frame made of pixels with $ r,c $ row and column coordinates 
* $ m $ (constant) Number of Lines in the sensor 

<br/>

Let's now build the $ t(u,i) $ *Pixel Timestamp Function* that associates to each pixel in $ u_{i} $ the appropriate timestamp 

In order to build this function, it is enough to compute the timestamp for each pixel according considering that 

* lines are read subsequently 
  * hence it is possible to think about the percentage of $ t_{\tau} $ total frame time consumed to read the first $ r $ lines over the $ m $ total number of lines 

then the final formula is as follows 

$$ t((r,c), i) = t_{i} + t_{\tau} \frac{r}{m} $$


## Transformation among frames 

In Rolling Shutter Cameras the effects of 

* self-motion 
* not static scene 

lead to distorsions such as 

<img src="http://www.andor.com/Portals/0/rolling-global2.jpg"/>

varying depending on the kind of motion 

Hence Transformations like for example Homography (holding when no translation is present) should be changed 

* In case of Pixels consistency (hence Global Shutter) they hold for all the Image 
* In case of Pixels non-consistency (hence Rolling Shutter) they hold for **each pixel separately**
  * This is because of Transformation for each pixel can be different 

Hence the Rotation Transformation 

$$ u_{i} = KR(t)u_{j} $$

holding between frame $ u_{i} $ and frame $ u_{j} $ has to be interpreted as follows 

* for Global Shutter 
  * $ R(t) $ is the same for all the pixels 
* for Rolling Shutter 
  * $ R(t) $ changes for every pixel hence 
    * it becomes $ R(t(r,c)) $ where $ r,c $ are the row and column coordinates of the pixel
